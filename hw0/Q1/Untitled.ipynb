```python
from torchvision import models
model = models.resnet101(pretrained=True)
#model.eval()
```

```python
from PIL import Image
from torchvision import transforms
img1 = Image.open('C:/Users/Shruthi/Documents/HWs/RTML/hw0/images/abacus.jpg')
img2 = Image.open('C:/Users/Shruthi/Documents/HWs/RTML/hw0/images/flamingo.jpg')
img3 = Image.open('C:/Users/Shruthi/Documents/HWs/RTML/hw0/images/rainbow.jpg')
img4 = Image.open('C:/Users/Shruthi/Documents/HWs/RTML/hw0/images/snowman.jpg')
img5 = Image.open('C:/Users/Shruthi/Documents/HWs/RTML/hw0/images/stocking.jpg')

preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
input_tensor1 = preprocess(img1)
input_tensor2 = preprocess(img2)
input_tensor3 = preprocess(img3)
input_tensor4 = preprocess(img4)
input_tensor5 = preprocess(img5)

input_batch1 = input_tensor1.unsqueeze(0) 
input_batch2 = input_tensor2.unsqueeze(0) 
input_batch3 = input_tensor3.unsqueeze(0) 
input_batch4 = input_tensor4.unsqueeze(0) 
input_batch5 = input_tensor5.unsqueeze(0) 
```

```python
output1 = model(input_batch1)
output2 = model(input_batch2)
output3 = model(input_batch3)
output4 = model(input_batch4)
output5 = model(input_batch5)
```

```python
import torch
probabilities1 = torch.nn.functional.softmax(output1[0], dim=0)
probabilities2 = torch.nn.functional.softmax(output2[0], dim=0)
probabilities3 = torch.nn.functional.softmax(output3[0], dim=0)
probabilities4 = torch.nn.functional.softmax(output4[0], dim=0)
probabilities5 = torch.nn.functional.softmax(output5[0], dim=0)

#print(probabilities)
```

```python
from PIL import Image

with open('C:/Users/Shruthi/Documents/HWs/RTML/hw0/images/imagenet_classes.txt', "r") as f:
    categories = [s.strip() for s in f.readlines()]
top5_prob1, top5_catid1 = torch.topk(probabilities1, 5)
top5_prob2, top5_catid2 = torch.topk(probabilities2, 5)
top5_prob3, top5_catid3 = torch.topk(probabilities3, 5)
top5_prob4, top5_catid4 = torch.topk(probabilities4, 5)
top5_prob5, top5_catid5 = torch.topk(probabilities5, 5)

print("Abacus.jpg \n")
print(categories[top5_catid1[0]],top5_prob1[0].item()*100,"%")
print ("\n")
for i in range(top5_prob1.size(0)):
    print(categories[top5_catid1[i]], top5_prob1[i].item()*100,"%")
print ("\n")   
print("Flamingo.jpg \n")
print(categories[top5_catid2[0]],top5_prob2[0].item()*100,"%")
print ("\n")
for i in range(top5_prob2.size(0)):
    print(categories[top5_catid2[i]], top5_prob2[i].item()*100,"%")
print ("\n")
print("Rainbow.jpg \n")
print(categories[top5_catid3[0]],top5_prob3[0].item()*100,"%")
print ("\n")
for i in range(top5_prob3.size(0)):
    print(categories[top5_catid3[i]], top5_prob3[i].item()*100,"%")
print ("\n")    
print("snowman.jpg \n")
print(categories[top5_catid4[0]],top5_prob4[0].item()*100,"%")
print ("\n")
for i in range(top5_prob4.size(0)):
    print(categories[top5_catid4[i]], top5_prob4[i].item()*100,"%")
print ("\n")    
print("Stocking.jpg \n")
print(categories[top5_catid5[0]],top5_prob5[0].item()*100,"%")
print ("\n")
for i in range(top5_prob5.size(0)):
    print(categories[top5_catid5[i]], top5_prob5[i].item()*100,"%")
```
